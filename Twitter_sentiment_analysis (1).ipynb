```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("twitter.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download("punkt")
from nltk.corpus import stopwords
nltk.download("stopwords")
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
```

```python
stp = stopwords.words("english")
```

```python
lemma = WordNetLemmatizer()
```

```python
def clean_text(text):
  text = word_tokenize(text.lower())
  na = [i for i in text if i.isalpha()]
  stpw = [i for i in na if i not in stp]
  l = [lemma.lemmatize(i,pos="v") for i in stpw]
  j = " ".join(l)
  return j
```

```python
df["new_tweet"] = df["tweet"].apply(clean_text)
```

```python
df.head()
```

```python
x = df["new_tweet"]
y = df["label"]
```

```python
blanks = []
for text in df["new_tweet"]:
  blanks.append(len(word_tokenize(text)))
```

```python
blanks
```

```python
df["sentence"] = blanks
```

```python
df.info()
```

```python
mid = (np.quantile(df.sentence,0.95))
```

```python
print(mid)
```

```python
import tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tok = Tokenizer()
tok.fit_on_texts(x)
tok.index_word
```

```python
vocb_clean = len(tok.index_word)
vocb_clean
```

```python
seq = tok.texts_to_sequences(x)
print(seq)
```

```python
from tensorflow.keras.preprocessing import sequence
```

```python
sequ = sequence.pad_sequences(seq,maxlen = int(mid))
```

```python
sequ
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(sequ,y,test_size=0.3,random_state=1)
```

LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN,LSTM,Dropout,Embedding
```

```python
rnn = Sequential()
```

```python
rnn.add(Embedding(vocb_clean+1,400,input_length = int(mid),mask_zero=True))
```

```python
rnn.add(LSTM(units=100,activation="tanh"))
rnn.add(Dropout(rate=0.2))
```

```python
rnn.add(Dense(units=50,activation="relu"))
rnn.add(Dropout(rate=0.2))
```

```python
rnn.add(Dense(units=1,activation="sigmoid"))
```

```python
rnn.compile(optimizer="adam",loss="binary_crossentropy", metrics = ["accuracy"])
```

```python
rnn.fit(xtrain,ytrain,batch_size=512,epochs=25)
```

```python
ypred = rnn.predict(xtest)
```

```python
ypred
```

```python
ypred = ypred>0.5
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred))
```
